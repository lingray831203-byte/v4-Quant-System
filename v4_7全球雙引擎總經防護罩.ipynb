```python
import yfinance as yf
import pandas as pd
from datetime import datetime
import warnings
warnings.filterwarnings('ignore') # 讓系統閉嘴，保持戰報純淨

# ==========================================
# 0. 🌍 宏觀總經警報設定 (Global Kill Switch)
# ==========================================
MACRO_SETTINGS = {
    "DXY_TICKER": "DX=F", # Yahoo Finance 的美元指數代號
    "DXY_DANGER_LEVEL": 105.0 # 風控長設定的危險水位 (大於此數值強制鎖死)
}

# ==========================================
# 1. 核心防禦陣型設定 (你的不講武德名單)
# ==========================================
DEFENSE_GRID_PLAN = {
    "0056.TW": {
        "name": "0056",
        "cost": 37.68,
        "grid_1": 36.50,
        "grid_2": 35.50,
        "msg_wait": "死守護城河，拒絕向上攤平"
    },
    "00878.TW": {
        "name": "00878",
        "cost": 22.50,
        "grid_1": 21.60,
        "grid_2": 21.00,
        "msg_wait": "流血不夠多，未達 21.60 散戶斷頭價不撿"
    },
    "2330.TW": {
        "name": "台積電",
        "cost": 1654,
        "grid_1": 1650,
        "grid_2": 1600,
        "msg_wait": "【絕對不准買】等外資砸破 1650 解開封印"
    },
    "2308.TW": {
        "name": "台達電",
        "cost": 850,
        "grid_1": 800,
        "grid_2": 750,
        "msg_wait": "不到 800 不彎腰，耐心等墜崖恐慌"
    }
}

# ==========================================
# 2. 獲取 DXY 美元指數狀態
# ==========================================
def get_dxy_status():
    try:
        df_dxy = yf.download(MACRO_SETTINGS["DXY_TICKER"], period="5d", interval="1d", progress=False)
        if not df_dxy.empty:
            current_dxy = float(df_dxy['Close'].iloc[-1])
            return current_dxy
    except:
        pass
    return None

# ==========================================
# 3. 帶有宏觀防護罩的網格核心
# ==========================================
def analyze_grid_defense(grid_plan):
    results = []

    # 啟動總經掃描
    current_dxy = get_dxy_status()
    global_lockdown = False

    if current_dxy:
        danger_level = MACRO_SETTINGS["DXY_DANGER_LEVEL"]
        results.append(f"🌍 【宏觀雷達】目前 DXY 美元指數: {current_dxy:.2f} (警戒線: {danger_level})")
        if current_dxy >= danger_level:
            global_lockdown = True
            results.append("🚨 【最高警報】美元極度強勢！外資瘋狂提款中！")
            results.append("🚨 【風控指令】觸發全局鎖死機制，強制關閉所有買進許可！\n")
        else:
            results.append("✅ 【總經綠燈】資金未見極端撤出，網格陷阱維持原定計畫。\n")
    else:
        results.append("⚠️ 無法取得美元指數，基於風控原則，請手動確認大盤風險。\n")

    # 執行個股掃描
    for sym, plan in grid_plan.items():
        try:
            df = yf.download(sym, period="1mo", interval="1d", progress=False)
            if df.empty:
                results.append(f"| {plan['name']:<5} | ⚠️ 報價連線中斷")
                continue

            current_price = float(df['Close'].iloc[-1])
            g1 = plan['grid_1']
            g2 = plan['grid_2']
            name = plan['name']
            custom_msg = plan['msg_wait']

            # 🔥 風控長嚴格紀律判定 (加入全局鎖死判斷) 🔥
            if global_lockdown:
                action = f"🛑 宏觀警報：全局鎖死，禁止接刀！"
            else:
                if current_price > g1:
                    action = f"🚫 {custom_msg}"
                elif current_price <= g2:
                    action = f"🚨 終極恐慌！觸發第二網 ({g2} 全力撿屍)"
                elif current_price <= g1:
                    action = f"⚠️ 破防！觸發第一網 ({g1} 開始收割)"
                else:
                    action = f"⏳ 系統監控中..."

            res_str = f"| {name:<5} | 價:{round(current_price,2):<7} | 第一網:{g1:<6} | 狀態: {action}"
            results.append(res_str)

        except Exception as e:
            results.append(f"| {plan['name']:<5} | ❌ 運算錯誤")

    return results

# ==========================================
# 4. 主程式：測試防禦矩陣
# ==========================================
if __name__ == "__main__":
    print(f"[{datetime.now().strftime('%H:%M:%S')}] 啟動 V4.7 終極防禦矩陣 (含 DXY 宏觀防護罩)...\n")

    defense_picked = analyze_grid_defense(DEFENSE_GRID_PLAN)

    print("=== 🛡️ 投資長核心防禦陣型 (左側價值網格) ===")
    print("-" * 90)
    if defense_picked:
        for p in defense_picked: print(p)
    print("-" * 90)
```
